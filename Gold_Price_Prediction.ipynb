## Model Building

* X = Feature Matrix (Independent Variable) 
* y = Target Variable (Dependent Variable) 

```python
X = data_df.drop(columns = ['GLD'], axis = 1)
y = data_df['GLD']

X
```

```python
y
```

### Data Standardization

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X)
```

```python
standardized_data = scaler.transform(X)
X = standardized_data
X
```

## Train-Test Split

```python
from sklearn.model_selection import train_test_split, RandomizedSearchCV
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=45)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
print(y.shape, y_train.shape, y_test.shape)
```

## Model Comparison

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

Evaluate the performance of different regression models on the given dataset. 

```python
#Below we will evaluate 6 different regression models
models = [LinearRegression, Lasso, Ridge, SVR, DecisionTreeRegressor, RandomForestRegressor]

mae_scores = []#Mean Absolute Error
mse_scores = []#Mean Squared Error
rmse_scores = []#Root Mean Squared Error
r2_scores = []#R-squared

#Below we will iterate over each model
for model in models:
    regressor = model().fit(X_train, y_train)
    y_pred = regressor.predict(X_test)
    
    mae_scores.append(mean_absolute_error(y_test, y_pred))
    mse_scores.append(mean_squared_error(y_test, y_pred))
    rmse_scores.append(mean_squared_error(y_test, y_pred, squared=False))
    r2_scores.append(r2_score(y_test, y_pred))
```

Let's create a pandas DataFrame to summarize and present the evaluation metrics of different regression models. 

```python
regression_metrics_df = pd.DataFrame({
    "Model": ["Linear Regression", "Lasso", "Ridge", "SVR", "Decision Tree Regressor", "Random Forest Regressor"],
    "Mean Absolute Error": mae_scores,
    "Mean Squared Error": mse_scores,
    "Root Mean Squared Error": rmse_scores,
    "R-squared (R2)": r2_scores
})

regression_metrics_df.set_index('Model', inplace=True)
regression_metrics_df
```

The Decision Tree Regressor and Random Forest Regressor have the lowest prediction error.
